```python
import torch
from torchvision.io import read_image
from src.bounding_box import BBoxModel
import json
import math

print(torch.cuda.get_device_name(0))
```

```python
MODEL_PATH = "./models/bbox_net_trained.pth"

batch_size = 16

image_dir = './data'
data_path = './data/NewData/accumulated_data/frame_data.json'
dest_path = './data/NewData/accumulated_data/frame_data_bbox_model.json'

model = BBoxModel(MODEL_PATH)
print(f"Using device: {model.device}")

with open(data_path, 'r') as f:
    data = dict(json.load(f))
```

```python
def load_image(img_name):
    image_path = f'{image_dir}/{data[img_name]["img_dir"]}/Images/{img_name}.png'
    img = read_image(image_path)
    return img
```

```python
batch_names = []
batch = []
for i, img_name in enumerate(data.keys()):
    batch_names.append(img_name)
    batch.append(load_image(img_name))

    if i % batch_size == batch_size-1 or i == len(data.keys())-1:
        predictions = model.find_bbox_many(batch)
        for j, pred in enumerate(predictions):
            data[batch_names[j]]['model_bbox'] = list(pred[0])
            data[batch_names[j]]['model_bbox'] = [float(tensor) for tensor in data[batch_names[j]]['model_bbox']]
        batch_names = []
        batch = []
        print(f'{math.ceil(i / batch_size)} batches out of {math.ceil(len(data.keys()) / batch_size)}')
```

```python
with open(dest_path, "w") as f:
    json.dump(data, f)
```
